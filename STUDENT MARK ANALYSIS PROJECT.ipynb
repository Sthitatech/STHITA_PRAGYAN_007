```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data_csv=pd.read_csv('Expanded_data_with_more_features.csv')
data_csv
```

```python
data_csv.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data_csv['WklyStudyHours']=data_csv['WklyStudyHours'].str.replace('05-Oct','5-10')
```

```python
data_csv.head()
```

```python
plt.figure(figsize=(5,5))
ax=sb.countplot(data=data_csv,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From the above chart,it can be analysed that:the number of female is more than male
```

```python
gb=data_csv.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
```

```python
gb
```

```python
sb.heatmap(gb, annot=True)
plt.show()
```

```python
# From the above chart,we have concluded that education of parents plays a vital role in chidrens's score.
```

```python
db=data_csv.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
```

```python
db
```

```python
sb.heatmap(db,annot=True)
```

```python
# From the above graph,it can be concluded that marital status of parents have a no/negligible impact on their child's score.
```

```python
groupA = data_csv.loc[(data_csv['EthnicGroup'] == 'group A')].count()
groupB = data_csv.loc[(data_csv['EthnicGroup'] == 'group B')].count()
groupC = data_csv.loc[(data_csv['EthnicGroup'] == 'group C')].count()
groupD = data_csv.loc[(data_csv['EthnicGroup'] == 'group D')].count()
groupE = data_csv.loc[(data_csv['EthnicGroup'] == 'group E')].count()
l=['group A','group B','group C','group D','group E']
mlist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mlist,labels=l,autopct='%1.2f%%')
plt.title('DISTRIBUTION OF ETHNIC GROUPS')
plt.show()
```
